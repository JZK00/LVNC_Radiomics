```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('./combine-1228.csv',encoding='gbk')
df.info()
```

```python
df.head()
```

```python
plt.figure(figsize=(20,20)) 
sns.heatmap(df.drop("End", axis=1).corr(),annot=True,cmap='Blues')
#plt.savefig("sns.jpg",bbox_inches = 'tight',dpi=1200) # save
plt.show()
```

```python
plt.figure(figsize=(20,20)) 
ax = sns.heatmap(df.drop("End", axis=1).corr(),annot=True,cmap='Blues',cbar=False)

label_y = ax.get_yticklabels()
plt.setp(label_y, rotation=0, horizontalalignment='right')
label_x = ax.get_xticklabels()
plt.setp(label_x, rotation=45, horizontalalignment='right')

#plt.savefig("sns1.jpg",bbox_inches = 'tight',dpi=1200)
plt.show()
```
